```python
import numpy as np
from sklearn.datasets import load_iris
from random import random
from sklearn.model_selection import train_test_split
import math
import operator
import pandas as pd
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
import math
import operator
```

```python
def euclid(x,y,n):
    distance = 0
    
    for i in range(n):
        distance += math.pow(x[i] - y[i], 2)
        
    return math.sqrt(distance)
```

```python
boston_dataset = load_boston()
X_train, X_test, y_train, y_test = train_test_split(boston_dataset['data'], boston_dataset['target'],test_size=0.1, random_state=23)
X_train.shape, X_test.shape

features=X_train.shape[1] 
elements=X_train.shape[0] 
tests=X_test.shape[0] 
k =3 ; 

for t in range(tests): 
    predicted=0;
    distance_array= []
    for j in range(elements):
        distance_array.append([j,euclid(X_test[t],X_train[j],features)]) 
                                                                         
    x=list(distance_array)
    x=sorted(x,key=operator.itemgetter(1)) 

    for i in range(k):
        predicted+=y_train[x[i][0]] 
    
    print(predicted/k,  y_test[t]) 
    
```

```python
boston_dataset = load_boston()
X_train, X_test, y_train, y_test = train_test_split(boston_dataset['data'], boston_dataset['target'],test_size=0.05, random_state=23)
X_train.shape, X_test.shape

features=X_train.shape[1] 
elements=X_train.shape[0] 
tests=X_test.shape[0] 
k =5 ; 

for t in range(tests): 
    predicted=0;
    distance_array= []
    for j in range(elements):
        distance_array.append([j,euclid(X_test[t],X_train[j],features)]) 
    x=list(distance_array)
    x=sorted(x,key=operator.itemgetter(1)) 
    division=0
    s=0.9 
    for i in range(k):
        s-=0.1
        division+=s
        predicted+=(s*y_train[x[i][0]]) 
    
    print(predicted/division,  y_test[t]) 
```
